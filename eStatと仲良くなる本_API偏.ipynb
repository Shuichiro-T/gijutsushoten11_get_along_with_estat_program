統計情報取得呼び出し例

```python
import requests

STAT_LIST_END_POINT = "http://api.e-stat.go.jp/rest/3.0/app/getStatsList"

def getStatsList(surveyYears=None,\
                 openYears=None,\
                 statsField=None,\
                 statsCode=None,\
                 searchWord=None,\
                 searchKind=None,\
                 collectArea=None,\
                 explanationGetFlg=None,\
                 statsNameList=None,\
                 startPosition=None,\
                 limit=None,\
                 updatedDate=None):
  
  url = STAT_LIST_END_POINT + "?appId=" + APP_ID + \
    ("&surveyYears=" + surveyYears if surveyYears is not None else "") + \
    ("&openYears=" + openYears if openYears is not None else "") + \
    ("&statsField=" + statsField if statsField is not None else "")  + \
    ("&statsCode=" + statsCode if statsCode is not None else "") + \
    ("&searchWord=" + searchWord if searchWord is not None else "") + \
    ("&searchKind=" + searchKind if searchKind is not None else "") + \
    ("&collectArea=" + collectArea if collectArea is not None else "") + \
    ("&explanationGetFlg=" + \
     explanationGetFlg if explanationGetFlg is not None else "") + \
    ("&statsNameList=" + statsNameList if statsNameList is not None else "") + \
    ("&startPosition=" + startPosition if startPosition is not None else "") + \
    ("&limit=" + limit if limit is not None else "") + \
    ("&updatedDate=" + updatedDate if updatedDate is not None else "") 
  
  result = requests.get(url)

  print("HTTPステータスコード:%s" % (result))

  print("レスポンス:\r\n%s" % (result.text))

getStatsList(statsCode="00200521", explanationGetFlg="N", limit="10", surveyYears="200001-200012")



```

統計データ取得呼び出し例

```python
import requests

STATS_DATA_END_POINT = "http://api.e-stat.go.jp/rest/3.0/app/getStatsData"

def getStatsData(dataSetId=None,\
                statsDataId=None,\
                lvTab=None,\
                cdTab=None,\
                cdTabFrom=None,\
                cdTabTo=None,\
                lvTime=None,\
                cdTime=None,\
                cdTimeFrom=None,\
                cdTimeTo=None,\
                lvArea=None,\
                cdArea=None,\
                cdAreaFrom=None,\
                cdAreaTo=None,\
                lvCat=None,\
                cdCat=None,\
                cdCatFrom=None,\
                cdCatTo=None,\
                startPosition=None,\
                limit=None,\
                metaGetFlg=None,\
                cntGetFlg=None,\
                explanationGetFlg=None,\
                annotationGetFlg=None,\
                replaceSpChar=None,\
                sectionHeaderFlg=None):
  
  url = STATS_DATA_END_POINT + "?appId=" + APP_ID + \
    ("&dataSetId=" + dataSetId if dataSetId is not None else "") + \
    ("&statsDataId=" + statsDataId if statsDataId is not None else "") + \
    ("&lvTab=" + lvTab if lvTab is not None else "") + \
    ("&cdTab=" + cdTab if cdTab is not None else "") + \
    ("&cdTabFrom=" + cdTabFrom if cdTabFrom is not None else "") + \
    ("&cdTabTo=" + cdTabTo if cdTabTo is not None else "") + \
    ("&lvTime=" + lvTime if lvTime is not None else "") + \
    ("&cdTime=" + cdTime if cdTime is not None else "") + \
    ("&cdTimeFrom=" + cdTimeFrom if cdTimeFrom is not None else "") + \
    ("&cdTimeTo=" + cdTimeTo if cdTimeTo is not None else "") + \
    ("&lvArea=" + lvArea if lvArea is not None else "") + \
    ("&cdArea=" + cdArea if cdArea is not None else "") + \
    ("&cdAreaFrom=" + cdAreaFrom if cdAreaFrom is not None else "") + \
    ("&cdAreaTo=" + cdAreaTo if cdAreaTo is not None else "") + \
    ("".join(
        ["&lvCat" + f"{i+1:02}" + "=" + lvCat[i] for i in range(0,len(lvCat))]
        )  if lvCat is not None else "") + \
    ("".join(
        ["&cdCat" + f"{i+1:02}" + "=" + cdCat[i] for i in range(0,len(cdCat))]
        )  if cdCat is not None else "") + \
    ("".join(
        ["&cdCat" + f"{i+1:02}" + "From=" + cdCatFrom[i] \
         for i in range(0,len(cdCatFrom))]
        )  if cdCatFrom is not None else "") + \
    ("".join(
        ["&cdCat" + f"{i+1:02}" + "To=" + cdCatTo[i] \
         for i in range(0,len(cdCatTo))]
        )  if cdCatTo is not None else "") + \
    ("&startPosition=" + startPosition if startPosition is not None else "") + \
    ("&limit=" + limit if limit is not None else "") + \
    ("&metaGetFlg=" + metaGetFlg if metaGetFlg is not None else "") + \
    ("&cntGetFlg=" + cntGetFlg if cntGetFlg is not None else "") + \
    ("&explanationGetFlg=" + \
     explanationGetFlg if explanationGetFlg is not None else "") + \
    ("&annotationGetFlg=" + \
     annotationGetFlg if annotationGetFlg is not None else "") + \
    ("&replaceSpChar=" + \
     replaceSpChar if replaceSpChar is not None else "") + \
    ("&sectionHeaderFlg=" + \
     sectionHeaderFlg if sectionHeaderFlg is not None else "") 
 
  result = requests.get(url)

  print("HTTPステータスコード:%s" % (result))

  print("レスポンス:\r\n%s" % (result.text))

getStatsData(statsDataId="0003404323",explanationGetFlg="Y")
```

データセット登録呼び出し例

```python
import requests

DATA_SET_END_POINT = "http://api.e-stat.go.jp/rest/3.0/app/postDataset"

def postDataset(dataSetId=None,\
                statsDataId=None,\
                lvTab=None,\
                cdTab=None,\
                cdTabFrom=None,\
                cdTabTo=None,\
                lvTime=None,\
                cdTime=None,\
                cdTimeFrom=None,\
                cdTimeTo=None,\
                lvArea=None,\
                cdArea=None,\
                cdAreaFrom=None,\
                cdAreaTo=None,\
                lvCat=None,\
                cdCat=None,\
                cdCatFrom=None,\
                cdCatTo=None,\
                openSpecified=None,\
                processMode=None,\
                dataSetName=None):
  
  url = DATA_SET_END_POINT + "?appId=" + APP_ID + \
    ("&dataSetId=" + dataSetId if dataSetId is not None else "") + \
    ("&statsDataId=" + statsDataId if statsDataId is not None else "") + \
    ("&lvTab=" + lvTab if lvTab is not None else "") + \
    ("&cdTab=" + cdTab if cdTab is not None else "") + \
    ("&cdTabFrom=" + cdTabFrom if cdTabFrom is not None else "") + \
    ("&cdTabTo=" + cdTabTo if cdTabTo is not None else "") + \
    ("&lvTime=" + lvTime if lvTime is not None else "") + \
    ("&cdTime=" + cdTime if cdTime is not None else "") + \
    ("&cdTimeFrom=" + cdTimeFrom if cdTimeFrom is not None else "") + \
    ("&cdTimeTo=" + cdTimeTo if cdTimeTo is not None else "") + \
    ("&lvArea=" + lvArea if lvArea is not None else "") + \
    ("&cdArea=" + cdArea if cdArea is not None else "") + \
    ("&cdAreaFrom=" + cdAreaFrom if cdAreaFrom is not None else "") + \
    ("&cdAreaTo=" + cdAreaTo if cdAreaTo is not None else "") + \
    ("".join(
        ["&lvCat" + f"{i+1:02}" + "=" + lvCat[i] for i in range(0,len(lvCat))]
        )  if lvCat is not None else "") + \
    ("".join(
        ["&cdCat" + f"{i+1:02}" + "=" + cdCat[i] for i in range(0,len(cdCat))]
        )  if cdCat is not None else "") + \
    ("".join(
        ["&cdCat" + f"{i+1:02}" + "From=" + cdCatFrom[i] \
         for i in range(0,len(cdCatFrom))]
        )  if cdCatFrom is not None else "") + \
    ("".join(
        ["&cdCat" + f"{i+1:02}" + "To=" + cdCatTo[i] \
         for i in range(0,len(cdCatTo))]
        )  if cdCatTo is not None else "") + \
    ("&openSpecified=" + openSpecified if openSpecified is not None else "") + \
    ("&processMode=" + processMode if processMode is not None else "") + \
    ("&dataSetName=" + dataSetName if dataSetName is not None else "")
 
  headers = {"Content-Type" : "application/x-www-form-urlencoded"}

  result = requests.post(url, headers=headers)

  print("HTTPステータスコード:%s" % (result))

  print("レスポンス:\r\n%s" % (result.text))

postDataset(statsDataId="0003404323",cdTab="020", cdCat=["100"], 
            cdCatFrom=["","","110"] , cdCatTo=["","","300"])
```

統計データ一括取得呼び出し例

```python
import requests
import json

STATS_DATAS_END_POINT = "http://api.e-stat.go.jp/rest/3.0/app/getStatsDatas"

# statsDatasSpec用のクラスJSONを作成する
class StatsDatasSpec:
  dataSetId=None
  statsDataId=None
  lvTab=None
  cdTab=None
  cdTabFrom=None
  cdTabTo=None
  lvTime=None
  cdTime=None
  cdTimeFrom=None
  cdTimeTo=None
  lvArea=None
  cdArea=None
  cdAreaFrom=None
  cdAreaTo=None
  lvCat=None
  cdCat=None
  cdCatFrom=None
  cdCatTo=None
  startPosition=None
  limit=None

  def str(self):
    value = {}
    if self.dataSetId is not None: value["dataSetId"] = self.dataSetId
    if self.statsDataId is not None: value["statsDataId"] = self.statsDataId
    if self.lvTab is not None: value["lvTab"] = self.lvTab
    if self.cdTab is not None: value["cdTab"] = self.cdTab
    if self.cdTabFrom is not None: value["cdTabFrom"] = self.cdTabFrom
    if self.cdTabTo is not None: value["cdTabTo"] = self.cdTabTo
    if self.lvTime is not None: value["lvTime"] = self.lvTime
    if self.cdTime is not None: value["cdTime"] = self.cdTime
    if self.cdTimeFrom is not None: value["cdTimeFrom"] = self.cdTimeFrom
    if self.cdTimeTo is not None: value["cdTimeTo"] = self.cdTimeTo
    if self.lvArea is not None: value["lvArea"] = self.lvArea
    if self.cdArea is not None: value["cdArea"] = self.cdArea
    if self.cdAreaFrom is not None: value["cdAreaFrom"] = self.cdAreaFrom
    if self.cdAreaTo is not None: value["cdAreaTo"] = self.cdAreaTo

    if self.lvCat is not None:
      for i in range(0,len(self.lvCat)):
        if self.lvCat[i] != "": value["lvCat" + f"{i+1:02}"] = self.lvCat[i]
    if self.cdCat is not None:
      for i in range(0,len(self.cdCat)):
        if self.cdCat[i] != "": value["cdCat" + f"{i+1:02}"] = self.cdCat[i]
    if self.cdCatFrom is not None:
      for i in range(0,len(self.cdCatFrom)):
        if self.cdCatFrom[i] != "": 
          value["cdCatFrom" + f"{i+1:02}"] = self.cdCatFrom[i]
    if self.cdCatTo is not None:
      for i in range(0,len(self.cdCatTo)):
        if self.cdCatTo[i] != "": 
          value["cdCatTo" + f"{i+1:02}"] = self.cdCatTo[i]
    if self.startPosition is not None: value["startPosition"] = self.startPosition
    if self.limit is not None: value["limit"] = self.limit
    
    return json.dumps(value).replace(" ", "")
    # return urllib.parse.urlencode(value) 

def getStatsDatas(metaGetFlg=None,\
                explanationGetFlg=None,\
                annotationGetFlg=None,\
                replaceSpChar=None,\
                sectionHeaderFlg=None,\
                statsDatasSpec=None):
  
  param = {}
  param["appId"] = APP_ID
  if metaGetFlg is not None: param["cdAreaTo"] = metaGetFlg
  if explanationGetFlg is not None: param["explanationGetFlg"] = explanationGetFlg
  if annotationGetFlg is not None: param["annotationGetFlg"] = annotationGetFlg
  if replaceSpChar is not None: param["replaceSpChar"] = replaceSpChar
  if sectionHeaderFlg is not None: param["sectionHeaderFlg"] = sectionHeaderFlg


  if statsDatasSpec is not None: 
    specStr = "[" 
    for spec in statsDatasSpec:
      specStr += spec + ","
    specStr = specStr.rstrip(",")
    specStr += "]"

    param["statsDatasSpec"] = specStr

  result = requests.post(STATS_DATAS_END_POINT,param )

  print("HTTPステータスコード:%s" % (result))

  print("レスポンス:\r\n%s" % (result.text))

# 一つ目の統計表
spec1 = StatsDatasSpec()
spec1.statsDataId = "0003404323"
spec1.cdTab = "020"
spec1.cdCat = ["100"]
spec1.cdCatFrom = ["","","110"]
spec1.cdCatTo = ["","","300"]

# 二つ目の統計表
spec2 = StatsDatasSpec()
spec2.statsDataId = "0003403679"
spec2.cdTab = "020"

getStatsDatas( statsDatasSpec=[spec1.str(), spec2.str()])
```
